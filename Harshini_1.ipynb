
This code segment aims to visualize the salary distribution for the top ten job titles in India. By analyzing the histogram, we gain insights into the range and frequency of salaries offered across various job roles. The visualization helps identify salary trends and disparities among different job titles, aiding in informed decision-making for job seekers and employers alike. The concise yet informative visualization facilitates a quick understanding of the salary landscape in India's data science industry.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Get top 10 job titles and company locations
top_10_job_titles = india_data['Job Title'].value_counts().nlargest(10).index
top_10_locations = india_data['Location'].value_counts().nlargest(10).index

# Histogram of Top 10 Salaries in India
plt.figure(figsize=(10, 6))
sns.histplot(data=india_data[india_data['Salary'].isin(india_data['Salary'].nlargest(10))], x='Salary', bins=30, kde=True, color='skyblue')
plt.title('Histogram of Top 10 Salaries in India')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()

```


This code segment presents a box plot illustrating the salary distribution for the top ten job titles in India's data science industry. The visualization offers insights into the salary range, median, and outliers for each job title, enabling a comparative analysis of salary levels across different roles. By succinctly presenting the salary distribution within the top job titles, the visualization aids in understanding the salary structure and potential earning opportunities in India's data science sector.

```python
# Box Plot of Top 10 Salaries by Top 10 Job Titles in India
plt.figure(figsize=(10, 6))
sns.boxplot(data=india_data[india_data['Job Title'].isin(top_10_job_titles)], x='Salary', y='Job Title', palette='viridis')
plt.title('Box Plot of Top 10 Salaries by Top 10 Job Titles in India')
plt.xlabel('Salary')
plt.ylabel('Job Title')
plt.show()


```


This code segment showcases a bar plot depicting the top ten locations in India and their corresponding average salaries in the data science domain. The visualization enables a quick comparison of salary levels across different geographical locations, aiding job seekers and employers in understanding regional salary trends. By presenting salary information in relation to specific locations, the visualization facilitates informed decision-making regarding job opportunities and business operations in India's data science industry.

```python
# Bar Plot of Top 10 Salaries by Top 10 Locations in India
plt.figure(figsize=(12, 8))
sns.barplot(data=india_data[india_data['Location'].isin(top_10_locations)], x='Salary', y='Location', ci=None, palette='coolwarm')
plt.title('Bar Plot of Top 10 Salaries by Top 10 Locations in India')
plt.xlabel('Salary')
plt.ylabel('Location')
plt.show()

```

```python
# Scatter Plot of Salaries vs. Years of Experience in India (for top 10 salaries)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=india_data[india_data['Salary'].isin(india_data['Salary'].nlargest(10))], x='Salaries Reported', y='Salary', color='orange')
plt.title('Scatter Plot of Top 10 Salaries vs. Salaries Reported in India')
plt.xlabel('Salaries Reported')
plt.ylabel('Salary')
plt.show()
```


This code segment generates side-by-side box plots to compare the salary distribution for the top ten job titles in both India and the USA's data science industries. The visualizations offer insights into the salary range, median, and outliers for each job title, facilitating a comparative analysis of salary levels across different roles. By presenting the salary distribution within the top job titles, the visualizations aid in understanding the salary structure and potential earning opportunities in both countries' data science sectors.

```python
# Get top 10 job titles for India and the USA
top_10_india_job_titles = india_data['Job Title'].value_counts().nlargest(10).index
top_10_us_job_titles = us_data['job_title'].value_counts().nlargest(10).index

# Filter data for top 10 job titles
india_top_jobs_data = india_data[india_data['Job Title'].isin(top_10_india_job_titles)]
us_top_jobs_data = us_data[us_data['job_title'].isin(top_10_us_job_titles)]

# Create box plots for top 10 job titles in India and the USA
plt.figure(figsize=(16, 8))

plt.subplot(1, 2, 1)  # Subplot for India
sns.boxplot(data=india_top_jobs_data, x='Salary', y='Job Title', palette='viridis')
plt.title('Salary Distribution for Top 10 Job Titles in India')
plt.xlabel('Salary')
plt.ylabel('Job Title')

plt.subplot(1, 2, 2)  # Subplot for the USA
sns.boxplot(data=us_top_jobs_data, x='salary', y='job_title', palette='magma')
plt.title('Salary Distribution for Top 10 Job Titles in the USA')
plt.xlabel('Salary')
plt.ylabel('Job Title')

plt.tight_layout()
plt.show()
```
